```python
# Re-import necessary modules after code execution state reset
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# Set up color scheme
effort_colors = {
    'Low Effort': '#E63946',     # Soft Red
    'Medium Effort': '#F4A261',  # Warm Orange
    'High Effort': '#2A9D8F'     # Teal Green
}

# Time span (50 years)
years = np.arange(0, 51)

# Define base efforts and increase rates (in %)
efforts_data = {
    'Reforestation': {
        'Low Effort': {'base': 0.2, 'rate': 2},
        'Medium Effort': {'base': 0.3, 'rate': 3},
        'High Effort': {'base': 0.4, 'rate': 4}
    },
    'Emission Reduction': {
        'Low Effort': {'base': 0.6, 'rate': 6},
        'Medium Effort': {'base': 0.7, 'rate': 7},
        'High Effort': {'base': 0.8, 'rate': 8}
    },
    'Buy Offset Credits': {
        'Low Effort': {'base': 0.4, 'rate': 4},
        'Medium Effort': {'base': 0.5, 'rate': 5},
        'High Effort': {'base': 0.6, 'rate': 6}
    }
}

sns.set_theme(style="darkgrid")

# Create figure and subplots
fig, axes = plt.subplots(1, 3, figsize=(12, 4), sharey=True)

for ax, (action, data) in zip(axes, efforts_data.items()):
    for level, values in data.items():
        base = values['base']
        rate = values['rate']
        # Calculate compounded effort over time
        effort = base * (1 + rate / 100) ** years
        ax.plot(years, effort, label=level, color=effort_colors[level])
    
    ax.set_title(f"{action} Effort Over 50 Years")
    ax.set_xlabel("Years")
    ax.grid(True)

axes[0].set_ylabel("Effort (%)")
axes[0].legend(title="Effort Level", loc="upper left")

# Save to PDF with proper layout
pdf_output_path = "./Effort_Comparison.pdf"
fig.tight_layout()
fig.savefig(pdf_output_path, format="pdf", bbox_inches="tight")

pdf_output_path
```
